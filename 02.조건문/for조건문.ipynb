```python
#for조건문 - for 변수 in 리스트(튜플 또는 문자열): 변수에 리스트 안의 값을 순서대로 대입하여 수행문장 실행
testlist = [1,2,3]
for i in testlist:
    print(i)
```

```python
#교재에 있는 예시인데 쥬피터에선 last가 정의가 안됬다고 뜸 근데 리스트안의 값이 튜플이라 변수를 알아서 잡는다고 설명되어 있는데 오류가 뜨는 이유를 잘 모르겠음 // 오타로 인한 오류 
a = [(1,2),(3,4),(5,6)]
for (first,last) in a:
    print(first+last)
```

```python
#5명의 학생이 시험을 보고 점수가 60점이상이면 합격 그렇지 않으면 불합격 결과표시
'''for a in s:
    if s >=60:
        print('success')
    else:
        print('failure')  이 구문은 오류발생 '>=' not supported between instances of 'list' and 'int' 
오류 발생한 이유는 대입할변수가 아닌 지정된 변수를 넣어서
✔ 왜 괄호로 묶으면 자동으로 튜플을 언패킹하는가?
for (first, last) in a:
이 구문은 다음처럼 언패킹(unpacking)을 수행함:
(1,2) → first=1, last=2
(3,4) → first=3, last=4
(5,6) → first=5, last=6
따라서 last가 “정의되지 않았다”고 뜨는 것이 아니라, **오타 때문에 다른 변수(lsat)**를 찾는 과정에서 생긴 에러.'''

s = [90,25,67,45,80]
n = 0
for a in s:
    n = n+1
    if a >= 60: 
        print('%d success' %n) 
    else:
        print('%d failure' %n)
        
```

```python
#교재에서 n을 넣은 이유는 번호를 매기기 위해서인거 같음
''' 작동구조는 a라는 변수에 s의 요소를 순차적으로 대입 n의 경우엔 번호를 붙여주기 위해서 한번 실행할때마다 +1이라는 표현식을 세움
그리고 요소 당 60점 이상이면 if를 아니라면 else의 결과를 표현하게끔 if 조건문 사용'''
s = [90,25,67,45,80]
n = 0
for a in s:
    if a >= 60: 
        print('%d success' %n) 
    else:
        print('%d failure' %n)
```

```python
#합격한 사람한테만 축하메세지를 보내고 불합격은 건너뛰는 구문
s = [90,25,67,45,80]
n = 0
for a in s:
    n = n+1
    if a <60:
        continue
    print('%d 번째 학생 합격 축하드립니다.' %n)
```

```python
#range함수 숫자리스트를 자동으로 만들어주는 함수 range(10) = 0~9 range(시작숫자,끝숫자) 시작숫자<=레인지범위<끝숫자
add = 0
for i in range(1,11):
    add = add + i
    print(add)
print(add)
```

```python
'''print(변수) 의 원래 형태는 변수+'\n' 인거임
문법	             정확한 의미
print(end=' ')	“print 후 줄바꿈 대신 공백을 붙여라”
print(end='')	“print 후 아무것도 안 붙여라 (줄도 안 바뀜)”
print('')	    “빈 문자열 출력 + 줄바꿈하여 다음 줄로 넘어가라”
print()	         print('')와 완전히 동일                       '''
for i in range(2,10):
    for j in range(1,10):
        print(i*j, end = ' ') #end는 한줄로 표시해라? 같음
    print('')                 #또 '' print를 하니 한줄이었던게 각 2단,3단 별로 줄이 나뉨
```

```python
#리스트 컴프리헨션
a = [1,2,3,4]
result = []
for num in a:
    result.append(num*3)

print(result)
```

```python
#리스트 컴프리헨션 [변수표현식 for 변수 in 리스트]
a = [1,2,3,4]
result = [num*3 for num in a]
print(result)
```

```python
#리스트 컴프리헨션 if추가 [변수표현식 for 변수 in 리스트 if 조건]
a = [1,2,3,4]
result = [num*3 for num in a if num % 2 == 0]
print(result)
```
