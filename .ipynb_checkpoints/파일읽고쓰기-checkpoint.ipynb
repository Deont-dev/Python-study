```python
#파일 생성하기 // 반드시 끝은 f.close()로 닫자. // w 쓰기, r 읽기, a 내용추가, b 바이너리모드
f = open("새파일.txt", 'w')
f.close()
```

```python
f = open("새파일.txt", 'w')
for i in range(1,11):
    data = "%d번째 줄입니다.\n" %i
    f.write(data)              # data를 파일 객체 f에 써라 라는 명령어
f.close()
```

```python
#파일 내용읽기 1.readline함수 2.readlines 함수 // 앞 두가지는 한줄씩 읽기때문에 줄끝에 줄바꿈문자가 있음! 3.read 함수 사용 4.f를 객체를 이용
f = open("새파일.txt", 'r')
line = f.readline()
print(line)
f.close()
```

```python
f = open("새파일.txt", 'r')
while True:
    line = f.readline()
    if not line : break
    print(line)
f.close()
```

```python
while True:
    data = input()
    if not data: break
    print(data)
```

```python
f = open("새파일.txt", 'r')
lines = f.readlines()
for line in lines:
    print(line)
f.close()
```

```python
#파일을 읽을때 줄바꿈\n 을 제거해서 읽는법 = strip() 함수사용
f = open("새파일.txt", 'r')
lines = f.readlines()
for line in lines:
    line = line.strip()
    print(line)
f.close()
```

```python
f = open("새파일.txt", 'r')
data = f.read()
print(data)
f.close()
```

```python
f = open("새파일.txt", 'r')
for line in f:
    print(line)
f.close()
```

```python
#파일 내용 추가하기
f = open("새파일.txt", 'a')
for i in range(11,20):
    data = "%d번째 줄입니다.\n" %i
    f.write(data)
f.close()
```

```python
# with문 = 항상 파일을 open하고 close하는 번거로움 줄이기 // 문법 안 문장들이 다 실행된 후 벗어나는 순간 파일 자동 닫힘
# with open(~~~) as f:
with open("새파일.txt", 'r') as f:
    line = f.readline()
    print(line)
```

```python
#모듈 사용하기 sys argv는 인수 추출 명령어[인수1,인수2 ....] ex)파일이름이 sys 라면 args0은 확장자까지 sys.py
import sys

args = sys.argv[1:]
for i in args:
    print(i)
```
