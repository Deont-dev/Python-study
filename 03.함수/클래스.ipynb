```python
#생성자 만들기 // 앞에서 setdata를 일일히 지정 안하고 간편하게 들어갈수있는 방법
# __init__(self, 인수1, 인수2, ...) 앞에선 객체에 클래스를 부여하고 다음 메서드에 입력하는 식이지만 생성자를 부여하면 클래스를 부여할때 값을 바로넣어 메서드에 대입되는 방식

class FourCal():
    def __init__(self, first, second):
        self.first = first
        self.second = second
    def setdata(self, first, second):
        self.first = first
        self.second = second   
    def add(self):
        result = self.first + self.second
        return result
    def sub(self):
        result = esult = self.first - self.second
        return result
    def mul(self):
        result = self.first * self.second
        return result
    def div(self):
        result = self.first / self.second
        return result

# a = FourCal() 오류발생 생성자가 호출되서 빈괄호가 아닌 바로 값을 넣으면 setdata 자동생성
a = FourCal(4,2)
a. first
a. div()
```

```python
#클래스 상속 / class 상속받을클래스(상속할클래스):
class FourCal():
    def __init__(self, first, second):
        self.first = first
        self.second = second
    def setdata(self, first, second):
        self.first = first
        self.second = second   
    def add(self):
        result = self.first + self.second
        return result
    def sub(self):
        result = esult = self.first - self.second
        return result
    def mul(self):
        result = self.first * self.second
        return result
    def div(self):
        result = self.first / self.second
        return result
        
class MoreFourCal(FourCal):
    pass
# 상속받은 클래스엔 별다른 내용이 없지만 FourCal을 상속했기 때문에 메서드들은 다 사용가능

a = MoreFourCal(4,2)
a.mul()
```

```python
class MoreFourCal(FourCal):
    def pow(self):
        result = self.first ** self.second
        return result

a = MoreFourCal(4,2)
a.pow()
```

```python
# 해결방법이 메서드 오버라이팅 상속받는 클래스에 부모 클래스에 있는 동일한 메서드를 덧씌우는과정 = 메서드 오버라이팅
class SafeFourCal(FourCal):
    def div(self):
        if self.second == 0:
            return 0
        else:
            self.first / self.second
a = SafeFourCal(4,0)
a.div()
```
