```python
#if문 전제조건이 참이면 if절을 거짓이면 else문을 실행 / if 변수 (비교연산자) : , else:
money = 2000
if money >= 3000:
    print('택시')
else:
    print('걷기')
```

```python
money = 2000
card = True
if money >=3000 or card:
    print('taxi')
else:
    print('walk')
```

```python
pocket = ['paper','cellphone','money']
if 'money' in pocket:
    print('taxi')
else:
    print('walk')
```

```python
# elif는 이전 if가 거짓인데 아직 모든조건에 대한 판별이 안끝나서 최종거짓까지 가기 전의 단계, 개수엔 제한 없음
# 예를 들면 돈이 있으면 택시를타고, 돈은 없지만 카드가 있다면 버스를타고, 둘 다 없으면 걸어가라 
pocket = ['paper','cellphone','money']
card = True
if 'money' not in pocket:
    print('taxi')
elif card:
    print('bus')
else:
    print('walk')
```
