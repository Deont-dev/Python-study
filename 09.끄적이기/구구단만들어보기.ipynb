```python
def sagu(n):
    result = []
    for i in range(1,10):
        result.append(n*i)
    return result
```

```python
result = sagu(4)
print(result)
```

```python
def gugu(n):
    return [n*i for i in range(1,10)]
print(gugu(3))
```
