```python
#포맷함수를 이용한 문자정렬 0은 들어갈 내용 / :<,:>,:^ 왼쪽 오른쪽 가운데 방향 / 10은 문자열길이
'''포매팅과 헷갈리면 기억해보자 포매팅은 %d,%s 이런식으로 들어갈 값이면 포맷함수는 {0}이 들어갈 값이다 그리고 포매팅은 %10s,%-10s 이것이 문자열 전체10개의 길이에서 붙이는거라고
한다면 포맷함수는 {0:>10} 굴러가는 내용은 똑같다 방향을 정하고 길이를 정하고 다만 포매팅같은경우엔 방향이 오른쪽은 양수 왼쪽은 음수일뿐'''
"{0:<10}" .format("hi")
"{0:>10}" .format("hi")
"{0:^10}" .format("hi")
```

```python
#포맷함수 공백채우기 방향세우기 중간에 채울값 넣기 ex. 방향이 :> 이거라면 중간사이에 k를 넣고싶다 :k>
"{0:=^10}" . format("hi")
```

```python
"{0:k^10}" . format("hi")
```

```python
#f문자열 포매팅 파이썬 3.6이상 형식 f'~~~~' " " 작은따옴표든 큰따옴표든 상관없음 // 표현식 지원가능 {age +1} 이런것도 가능함
name = "홍길동"
age = 30
f"나의 이름은 {name}입니다. 나이는 {age}입니다."
```

```python
#딕셔너리를 활용하여 f문자열포매팅 이용
d = {'name' : '홍길동', 'age' : 30}
f'나의 이름은 {d["name"]}입니다. 나이는 {d["age"]}입니다.'
```

```python
#f문자열포매팅의 정렬방법 이것도 괄호안의 기본구조는 동일 {들어갈문자:채워넣을문자/방향/총 길이} 방향의 기본 디폴트값은 문자왼쪽,숫자오른쪽
f'{"hi":>10}'
```
