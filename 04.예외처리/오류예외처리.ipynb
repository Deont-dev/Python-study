```python
# try-except문 발생오류가 되면 해당 except문 실행
'''
try:
    ...
except [발생오류 [as 오류변수]]:
    ... '''

try:
    4/0
except ZeroDivisionError as e:
    print(e)
```

```python
# try-finally문 try문 수행 도중 예외 발생 여부에 상관없이 항상 수행된다. 사용한 리소스를 close해야할때 많이 사용
try:
    f = open('새파일.txt','r') # 수행되는 내용
finally:
    f.close() # 중간에 오류가 발생하더라도 무조건 실행
```

```python
# 여러개의 오류 처리 / 오류자체에 대해 진행
try:
    a = [1,2]
    print(a[3])
    4/0
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")
except IndexError:
    print("인덱싱 할 수 없습니다.")
```

```python
# 여러개의 오류 처리 / 변수로 표현
try:
    a = [1,2]
    print(a[3])
    4/0
except ZeroDivisionError as e:
    print(e)
except IndexError as e:
    print(e)
```

```python
# 여러개의 오류 처리 / 괄호로 묶어 한꺼번에 처리
try:
    a = [1,2]
    print(a[3])
    4/0
except (ZeroDivisionError,IndexError) as e:
    print(e)
```

```python
# try-else문 수행 중 오류가 발생하면 except절, 발생하지 않으면 else절
try:
    age = int(input("나이를 입력하세요: "))
except:
    print('입력이 정확하지 않습니다.')
else:
    if age <= 18:
        print('미성년자는 출입금지입니다.')
    else:
        print('환영합니다.')
```

```python
# 오류 회피하는법
try:
    f = open('나없는파일','r')
except FileNotFoundError:    # 파일이 없더라도 오류가 발생하지 않고 통과
    pass
```

```python
# 오류 일부러 발생시키기 / raise 명령어 사용 / 예시로 bird 클래스를 상속받는 자식 클래스는 반드시 fly라는 함수를 구현하도록 만들고 싶을때
class Bird:
    def fly(self):
        raise NotImplementedError

class Eagle(Bird):
    pass
eagle = Eagle()
eagle.fly()
```

```python
#Bird클래스에 이미 fly함수를 오류가 뜨게했으니 Eagle에서 fly메서드를 오버라이팅하지않으면 오류 발생하고 오버라이팅하면 한대로 출력
class Eagle(Bird):
    def fly(self):
        print('very speed')

eagle = Eagle() 
eagle.fly()
```

```python
# 예외 만들기 / 내장클래스 Exception 이용
class MyError(Exception):
    pass

def say_nick(nick):
    if nick == '바보':
        raise MyError()
    print(nick)

say_nick('천사')
say_nick('바보')
```

```python
# 예외처리기법으로 MyError를 예외처리해보자
try:
    say_nick('천사')
    say_nick('바보')
except MyError:
    print("허용되지 않는 별명입니다.")
```

```python
# 오류메세지 사용하고싶은 경우라면
try:
    say_nick('천사')
    say_nick('바보')
except MyError as e:
    print(e)
```

```python
#오류메세지가 안보인다? 오류 클래스에 __str__메서드를 구현해야됨
class MyError(Exception):
    def __str__(self):
        return '허용되지 않은 별명입니다.'

try:
    say_nick('천사')
    say_nick('바보')
except MyError as e:
    print(e)
```
